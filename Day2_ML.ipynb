```python
#Seaborn and matplotlib combinly

#tips dataset also available in default in python programming
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

```python
#We are creating bar graph for iris

plt.bar(data['species'], data['petal_length'])
plt.bar(data['species'], data['petal_width'])

plt.show()
```

```python
plt.bar(tips["day"], tips['total_bill'])
```

```python
plt.barh(tips["day"], tips['total_bill'])
```

```python
# Histogram

plt.hist(tips['total_bill'], edgecolor = "black")

plt.show()
```

```python
#Day must be in fixed length

day = tips.groupby('day')['total_bill'].sum().reset_index()

day
```

```python
#Pie chart

x = day['day']

y = day['total_bill']
```

```python
explode = [0,0,0,0.1]

plt.pie(y,labels=x,explode = explode,autopct='%1.1f%%',colors=['red','blue','green','yellow'],
        wedgeprops = {'edgecolor' : 'black'},shadow=True,startangle=90)
plt.show()
```

```python
#Scatter plot

plt.scatter(tips['total_bill'], tips['tip'])
plt.title("Scatter plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python

tips['size'].value_counts()
```

```python
#I want to create scatter plot with some color palet

#Scatter plot

plt.scatter(tips['total_bill'], tips['tip'],  s = 100, c = tips['size'])
plt.title("Scatter plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
cbar = plt.colorbar()
cbar.set_label("Size")
plt.show()
```

```python
#Scatterplot using seaborn

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette= 'summer')
```

```python
#Swarmplot

sns.swarmplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
#Strip plot

sns.stripplot(x = 'day', y = 'tip', data = tips, hue = 'sex')
```

```python
#Voiline Plot

sns.violinplot(x = 'day', y = 'tip', data = tips, hue = 'sex', split = True)
```

```python
#Boxplot

sns.boxplot(x = 'day', y = 'tip', data = tips, hue = 'sex')
```
